```python
# Dependencies and Setup
import pandas as pd
import os
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
num_schools = len(school_data)                   #number of schools
num_students = len(student_data)                 #number of students
total_budget = school_data['budget'].sum()       #total district budget

#average math score, sum of math score/num students
avg_math = (student_data['math_score'].sum() / num_students) 
avg_math = round(avg_math,2)

#average readign score, sum of read/num students
avg_read = (student_data['reading_score'].sum() / num_students)
avg_read = round(avg_read, 2)

#calculate students with a math score of 70 and over
math_seventy = (len(student_data[student_data.math_score >= 70]) / num_students) *100
math_seventy = round(math_seventy, 2)

#calculate 
read_seventy = (len(student_data[student_data.reading_score >=70])/ num_students) *100
read_seventy = round(read_seventy, 2)

#calculate percent student with math & reading score 70 and over
math_read_seventy = (len(student_data[(student_data.math_score >=70) & 
                                     (student_data.reading_score >=70)]) /num_students) *100

math_read_seventy = round(math_read_seventy, 2)

#create new dataframe for the district results 
results_dict = {'Number of Schools': [num_schools], 'Number of Students':[num_students], 'Total Budget':[total_budget],
              'Avg Math Score':[avg_math], 'Avg Reading Score':[avg_read], 'Passing Math %':[math_seventy], 
              'Passing Reading':[read_seventy], 'Passing Reading/Math':[math_read_seventy]}

results_df = pd.DataFrame (results_dict)
results_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_groupings = student_data.groupby('grade')
ninth_graders = student_data[(student_data["grade"] == "9th")]
ninth_graders_scores = round(ninth_graders.groupby(["school_name"]).mean()["math_score"],2)

tenth_graders = student_data[(student_data["grade"] == "10th")]
tenth_graders_scores = round(tenth_graders.groupby(["school_name"]).mean()["math_score"],2)

elvnth_graders = student_data[(student_data["grade"] == "11th")]
elvnth_graders_scores = round(elvnth_graders.groupby(["school_name"]).mean()["math_score"],2)

twlth_graders = student_data[(student_data["grade"] == "12th")]
twlth_graders_scores = round(twlth_graders.groupby(["school_name"]).mean()["math_score"],2)

math_scores_dict = {'9th Grade':ninth_graders_scores,'10th Grade':tenth_graders_scores, 
                    '11th Grade':elvnth_graders_scores, '12th Grade':twlth_graders_scores}

math_scores_df = pd.DataFrame(math_scores_dict)
math_scores_df.index.names=['School Name']

math_scores_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 584, 629, 644, 675]
labelsx = ["$575", '$576 - $624','$625 - $674', '>$725']
school_spending_df = school_summary_df

school_spending_df['Spending Ranges'] = pd.cut(school_spending_df['Per Student Budget'], spending_bins, labels=labelsx)
school_spending_df = round (school_spending_df.groupby("Spending Ranges").mean(),2)

school_spending_df = school_spending_df[['reading_score_x', 'math_score_x', 'math_score_y', 'reading_score_y', 'grade']]

school_spending_df = school_spending_df.rename(columns ={"reading_score_x": "Average Reading Score", 
                                                        "math_score_x":"Average Math Score","math_score_y":"% Passing Math", "reading_score_y":"% Passing Reading", 
                               "grade":"% Overall Passing"})

school_spending_df.sort_values(["% Overall Passing"], ascending = False)
```
